```python
TrainCols = list(df1.columns.values)
TestCols = list(df2.columns.values)
print(TrainCols)
print(TestCols)
```

```python
Xtrain = df1[TrainCols[0:len(TrainCols)-1]].copy()
Ytrain = df1[['TARGET']].copy()
print(Xtrain.shape)
print(Ytrain.shape)
Xtest = df2.copy()
```

```python
Xtest.shape
```

```python
import numpy as np
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix,classification_report
from sklearn.model_selection import train_test_split 
from sklearn import metrics 
from sklearn import tree
```

```python
dt=DecisionTreeClassifier()
dt.fit(Xtrain, Ytrain)
Y_Pred = dt.predict(Xtest)
Y_Pred = pd.DataFrame(Y_Pred,columns=['TARGET'])
Y_Pred.to_csv(index=False)
```

```python
clf = DecisionTreeClassifier(max_depth = None,random_state = 1, criterion= "gini")
clf.fit(Xtrain, Ytrain)

tree.plot_tree(clf);
```

```python
print("Count of 0 & 1 in TARGET column for Train Data")
print(Y_Pred['TARGET'].value_counts())
```

```python
X_Pred = dt.predict(Xtrain)
print("Accuracy:", metrics.accuracy_score(Ytrain,X_Pred))
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(Xtrain, Ytrain, test_size = .30, random_state = 0)
dt = dt.fit(X_train, Y_train)
Y_PredNew = dt.predict(X_test)
print("Accuracy:", metrics.accuracy_score(Y_test,Y_PredNew))
```

```python
print(metrics.confusion_matrix(Y_test, Y_PredNew))
print()
print('Printing the precision,recall,f1,support')
print(metrics.classification_report(Y_test, Y_PredNew))
```
